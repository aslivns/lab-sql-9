```python
# Create a Python connection with SQL database and retrieve the results of the following queries as dataframes:
import pandas as pd
import numpy as np

import pymysql     # for getting data from SQL database
from sqlalchemy import create_engine  #for establishing

from getpass import getpass  # To get the password without showing the input
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
# 1. How many distinct (different) actors' last names are there?

data = pd.read_sql_query("SELECT COUNT(DISTINCT last_name) AS distinct_last_name \
FROM sakila.actor;", engine)
data
```

```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
data1 = pd.read_sql_query("SELECT *, \
CASE WHEN DAYNAME(rental_date) IN ('Saturday', 'Sunday') THEN 'Weekend'\
ELSE 'Weekday' END AS day_type \
FROM rental;", engine)
data1.head()
```

```python
# 3. Get all films with ARMAGEDDON in the title.
data2 = pd.read_sql_query("SELECT title \
FROM film \
WHERE title LIKE ('%%ARMAGEDDON%%');", engine)
data2
```

```python
# 4. Get 10 the longest films.
data3 = pd.read_sql_query("SELECT title, length \
FROM film \
ORDER BY length DESC \
LIMIT 10; \
", engine)
data3
```

```python
# 5. How many films include Behind the Scenes content?
data4 = pd.read_sql_query("SELECT COUNT(*) AS behind_the_scenes_count \
FROM film \
WHERE special_features LIKE '%%Behind the Scenes%%';", engine)
data4.head()
```

```python
# 6.  Which kind of movies (rating) have a mean duration of more than two hours?
data5 = pd.read_sql_query(" SELECT rating, ROUND(AVG(length),2) AS avg_per_rating \
FROM film \
GROUP BY rating HAVING AVG(length) > 120 ;", engine)
data5
```

```python
# 7.  Rank films by length (filter out the rows that have nulls or 0s in length column). 
# In your output, only select the columns title, length, and the rank.

data6 = pd.read_sql_query("SELECT title, length, DENSE_RANK () OVER (ORDER BY length DESC) AS the_rank \
FROM film \
WHERE length !=0 AND length IS NOT NULL;", engine)
data6


## or can be dome with this query; 
# SELECT title, length,
# CASE
# WHEN length BETWEEN 1 AND 60 THEN 'Short'
# WHEN length BETWEEN 61 AND 120 THEN 'Medium'
# ELSE 'Long'
# END AS the_rank
# FROM film
# WHERE length IS NOT NULL AND length != 0
# ORDER BY length DESC;
```
